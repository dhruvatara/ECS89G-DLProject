```python
import numpy as np
import pandas as pd
df = pd.read_csv('somewhat_cleaned.csv')
twentycentury = df[df['Year']>1899]
```

```python
# considering only 20th century+ data because GDP and inflation values are NaN before
twentycentury.isna().sum()
```

```python
# canada = twentycentury[twentycentury['Country']=='Canada']
```

```python
country = twentycentury['Country'].tolist()
print(country)
```

```python
X = twentycentury.drop(['Unnamed: 0','Crisis','GDP PerCapita','Year','Country'],axis=1)
y = twentycentury['Crisis']
columns = X.columns
```

```python
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(X,columns=columns)
X.head()
```

```python
X["Countries"] = country
```

```python
X
```

```python
X_train = X.loc[X['Countries'] != 'United States']
```

```python
X_test = X.loc[X['Countries'] == 'United States']
```

```python
X_train = X_train.drop(['Countries'], axis=1)
X_train.shape
```

```python
X_test = X_test.drop(['Countries'], axis=1)
X_test.shape
```

```python
y_train = y[:232]
print(y_train.shape)
```

```python
y_test = y[-116:]
print(y_test.shape)
```

```python
# Buid RNN model
# Import libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential()

model.add(layers.Embedding(input_dim=2, output_dim=32)) #input_dim to be decided...

# Add a LSTM layer with 32 internal units.
model.add(layers.LSTM(32,activation='sigmoid', recurrent_activation='sigmoid'))

# Add a Dense layer with 1 unit.
model.add(layers.Dense(1,activation='sigmoid'))

history = model.compile(optimizer="Adam", loss="binary_crossentropy", metrics=[tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

model.summary()
```

```python
model.fit(X_train,y_train,
         epochs=20,
         batch_size=128)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict = y_predict.reshape(1, 116)
y_predict = y_predict[0].tolist()

y_test = y_test.tolist()
```

```python
for i in range(len(y_predict)):
    print("True crisis data: {} <=> Predicted crisis data: {}".format(y_test[i], y_predict[i]))
```
